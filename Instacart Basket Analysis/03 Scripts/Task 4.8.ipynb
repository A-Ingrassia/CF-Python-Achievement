### 3) Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Splits the data into groups based on the 'user_id' column.
# Applies the transform() function on the 'order_number column to generate the max orders for each user.
# Creates a 'max_order' to be populated with the results of the aggregation.

ords_prods_merged['max_order'] = ords_prods_merged.groupby('user_id')['order_number'].transform(np.max)
```

```python
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”

ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# If the max orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled 
# a “Regular customer.”

ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Prints the frequency of the new 'loyalty_flag' column using the value_counts() function

ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head()
```

### 5) The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

    A) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    
    B) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them 
    as a “High spender.”


```python
# Creates a new column ('average_price') that will contain the average price of each user's cumulative orders

ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Observing the new column

ords_prods_merged.head()
```

```python
# Creates a 'spending_category' flag for low spenders

ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_category'] = 'Low spender'
```

```python
# Creates a 'spending_category' flag for high spenders

ords_prods_merged.loc[ords_prods_merged['average_price'] > 10, 'spending_category'] = 'High spender'
```

```python
# Observing the new column

ords_prods_merged.head()
```

```python
# Observing the frequency of each flag

ords_prods_merged['spending_category'].value_counts(dropna = False)
```

### 6) In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

    A) If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a 
     “Non-frequent customer.”
     
    B) If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular 
    customer.”
    
    C) If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Creates the 'median_order_frequency' column using transform()

ords_prods_merged['median_order_frequency'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Observing the new column

ords_prods_merged.head()
```

```python
# Creates an 'order_frequency' flag for non-frequent customers

ords_prods_merged.loc[ords_prods_merged['median_order_frequency'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
# Creates an 'order_frequency' flag for regular customers

ords_prods_merged.loc[(ords_prods_merged['median_order_frequency'] > 10) & (ords_prods_merged['median_order_frequency'] <= 20), 'order_frequency'] = 'Regular customer'
```

```python
# Creates an 'order_frequency' flag for frequent customers

ords_prods_merged.loc[ords_prods_merged['median_order_frequency'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Observing the flags

ords_prods_merged.head()
```

```python
# Observing the frequency of each flag

ords_prods_merged['order_frequency'].value_counts(dropna = False)
```
